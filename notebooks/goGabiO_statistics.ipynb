```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
# Input 
csv_file = "output_Air Bud Seventh Inning Fetch (2002).csv"
csv_file_ace_ventura = "https://raw.githubusercontent.com/moe221/diversity_in_cinema/master/results/ace_ventura_results.csv"
csv_file_A_team = "output_The A-Team (2010).csv"
```

```python
df_movie = pd.read_csv(csv_file_ace_ventura)
df_movie.head(20)
```

```python
df_movie[df_movie['race']=='middle eastern']
```

```python
genders = pd.unique(df_movie['gender'])
genders
```

```python
races = pd.unique(df_movie['race'])
races
```

```python
df_movie_2 = df_movie.copy()
```

```python
ohe_g = OneHotEncoder(sparse = False)

ohe_g.fit(df_movie_2[['gender']])

gender_encoded = ohe_g.transform(df_movie_2[['gender']])
```

```python
ohe_g.categories_[0][0]
```

```python
#use ohe.categories_ in for loop here
results_g = gender_encoded.T

for i, cat in enumerate(ohe_g.categories_[0]):
    df_movie_2[cat] = results_g[i]

df_movie_2.head()
```

```python
ohe_r = OneHotEncoder(sparse = False)

ohe_r.fit(df_movie_2[['race']])

race_encoded = ohe_r.transform(df_movie_2[['race']])

#use ohe.categories_ in for loop here
results_r = race_encoded.T

for i, cat in enumerate(ohe_r.categories_[0]):
    df_movie_2[cat] = results_r[i]

df_movie_2.head()
```

for index, row in df_movie_2.iterrows():
    if row['gender'] == 'woman' and row['race'] == 'white':
        row['women_of_color'] = 1

```python
df_movie_2['women_of_color'] = df_movie_2['gender'] + ' ' + df_movie_2['race']
```

```python
def woman_of_color(x):
    if 'Woman' in x and 'white' not in x:
        return 1
    return 0
```

```python
df_movie_2['women_of_color'] = df_movie_2['women_of_color'].apply(woman_of_color)
```

```python
df_movie_2['women_of_color'].unique()
```

```python
df_movie_2['women_of_color'].sum()/len(df_movie_2)
```

```python
df_movie_2.head()
```

```python
df_movie_3 = df_movie_2.groupby('frame_number').sum()
df_movie_3['face_count'] = df_movie_3['Man'] + df_movie_3['Woman']
```

```python
df_movie_3.head()
```

```python
only_men = len(df_movie_3[df_movie_3['Woman']==0])
only_women = len(df_movie_3[df_movie_3['Man']==0])
```

```python
dict_stats = {'total_frames': [len(df_movie_3)], 'total_seconds': [len(df_movie_3)/2],
              'total_faces': [df_movie_3['face_count'].sum()],
              'total_men': [df_movie_3['Man'].sum()],             
              'total_women': [df_movie_3['Woman'].sum()],
              'total_asian': [df_movie_3['asian'].sum()],
              'total_black': [df_movie_3['black'].sum()],
              'total_indian': [df_movie_3['indian'].sum()],
              'total_latino_hispanic': [df_movie_3['latino hispanic'].sum()],
              'total_middle_eastern': [df_movie_3['middle eastern'].sum()],
              'total_white': [df_movie_3['white'].sum()],
              'total_women_of_color': [df_movie_3['women_of_color'].sum()],
              'only_men': only_men, 'only_women': only_women}
```

```python
dict_stats
```

```python
df_stats = pd.DataFrame.from_dict(dict_stats)
```

```python
df_stats
```

```python
# gender screentime

man_screentime = df_stats['total_men']/df_stats['total_faces']*100
woman_screentime = df_stats['total_women']/df_stats['total_faces']*100
only_men = df_stats['only_men']/df_stats['total_frames']*100
only_women = df_stats['only_women']/df_stats['total_frames']*100

print(man_screentime)
print(woman_screentime)
print(only_men)
print(only_women)
```

```python
# race screentime

asian_screentime = df_stats['total_asian']/df_stats['total_faces']*100
black_screentime = df_stats['total_black']/df_stats['total_faces']*100
indian_screentime = df_stats['total_indian']/df_stats['total_faces']*100
latino_hispanic_screentime = df_stats['total_latino_hispanic']/df_stats['total_faces']*100
middle_eastern_screentime = df_stats['total_middle_eastern']/df_stats['total_faces']*100
white_screentime = df_stats['total_white']/df_stats['total_faces']*100
print(asian_screentime)
print(black_screentime)
print(indian_screentime)
print(latino_hispanic_screentime)
print(middle_eastern_screentime)
print(white_screentime)
print(asian_screentime+black_screentime+indian_screentime+latino_hispanic_screentime+middle_eastern_screentime+white_screentime)
```
